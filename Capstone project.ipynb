```python
data = df['partyname'].value_counts(ascending=False)
data
datalessthan700 = data[data < 700]
datalessthan700

def partname(value):
    if value in datalessthan700:
        return 'others'
    else:
        return value

df['partyname'] = df['partyname'].apply(partname)
df['partyname'].value_counts()
```

```python
df.replace(to_replace=['Independents'], value ='Independent', inplace=True)
df.replace(to_replace=['IND'], value ='Independent', inplace=True)
df['partyname'].replace(to_replace=['BJP'], value ='Bharatiya Janata Party', inplace=True)
df['partyname'].replace(to_replace=['BSP'], value ='Bahujan Samaj Party', inplace=True)
df['partyname'].replace(to_replace=['INC'], value ='Indian National Congress', inplace=True)
```

```python
df['partyname'].value_counts()
```

```python
dataparty = df['partyabbre'].value_counts()
```

```python
partyabbre700 = dataparty[dataparty < 700]
```

```python
def partyabbre(value):
    if value in partyabbre700:
        return 'others'
    else:
        return value

df['partyabbre'] = df['partyabbre'].apply(partyabbre)
df['partyabbre'].value_counts()
```
